```python
def drop_unwanted_columns(df):
    columns_to_drop = [
        'present_use', 'land_val', 'imp_val', 'year_built', 'year_reno',
        'sqft_lot', 'sqft', 'sqft_1', 'sqft_fbsmt', 'grade', 'fbsmt_grade',
        'condition', 'stories', 'beds', 'bath_full', 'bath_3qtr',
        'garb_sqft', 'gara_sqft', 'view_olympics', 'view_cascades',
        'view_territorial', 'view_skyline', 'view_sound', 'view_lakewash',
        'view_lakesamm', 'view_otherwater', 'view_other','sale_date'
    ]
    return df.drop(columns=columns_to_drop, errors='ignore')
```

```python
df.drop(['noise_traffic'],axis=1,inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python

le = LabelEncoder()
for col in ['city', 'zoning', 'subdivision', 'submarket', 'sale_warning', 'join_status']:
    df[col] = le.fit_transform(df[col].astype(str)) 
```

```python
df.info()
```

```python
df['sale_date'] = pd.to_datetime(df['sale_date'])
```

```python

from sklearn.metrics import r2_score
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['sale_price','sale_date'] , axis=1)
y = df['sale_price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = RandomForestRegressor()
model.fit(x_train, y_train)
```

```python
pred_1 = model.predict(x_test)
accu = r2_score(y_test,pred_1)
```
