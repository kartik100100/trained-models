```python
import pandas as pd 
import numpy as np 
import seaborn as sns
```

```python
ds = pd.read_csv(r"C:\Users\karti\Downloads\train.csv")
test = pd.read_csv(r"C:\Users\karti\Downloads\test (1).csv")
test_ids = test['id']
ds.head()
```

```python
for col in ['Social_event_attendance','Time_spent_Alone','Going_outside','Friends_circle_size','Post_frequency']:
    ds[col] = ds[col].fillna(ds[col].mean())
```

```python
ds.isnull().sum()
```

```python
for col in ['Drained_after_socializing','Social_event_attendance']:
    ds[col] = ds[col].fillna(ds[col].mode()[0])
```

```python
ds.isnull().sum()
```

```python
ds['Stage_fear']=ds['Stage_fear'].fillna(ds['Stage_fear'].mode()[0])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
ds['Drained_after_socializing'] =le.fit_transform(ds['Drained_after_socializing'] )
```

```python
ds['Personality']=le.fit_transform(ds['Personality'])
```

```python
ds['Stage_fear']=le.fit_transform(ds['Stage_fear'])
```

```python
ds.info()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
x = ds.drop('Personality' , axis=1)
y = ds['Personality']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
trf = FunctionTransformer()
```

```python
x_train1 = trf.fit_transform(x_train)
x_test1 = trf.fit_transform(x_test)
```

```python
model = DecisionTreeRegressor()
model.fit(x_train1,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
pred = model.predict(x_test)
accu = accuracy_score(y_test,pred)
print(accu)
```

```python
model_1 = DecisionTreeClassifier()
model_1.fit(x_train1,y_train)
```

```python
pred_1 = model_1.predict(x_test)
accu_1 = accuracy_score(y_test,pred_1)
print(accu_1)
```
